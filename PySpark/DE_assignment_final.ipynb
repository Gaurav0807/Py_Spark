# Moderate Questions

## Create a new dataframe(df_survived) which follow all the following condition at the same time from the original dataset
#### 1. All passengers should be male
#### 2. The age of each passenger should be greater than 40 but less than 50
#### 3. Passenger should not belong to 3rd PClass( Pclass not equal to 3)



```python
# 'Sex': df1['Sex']=='male',
#     'Age':df1[(df1['Age']>40) and (df1['Age']<50)],
#     'PClass'df1['Pclass']!=3
    
df10 = df1[(df1['Sex']=='male') & (df1['Pclass']!=3) & (df1['Age']>40) & (df1['Age']<50)]
df_survived=pd.DataFrame(df10)
df_survived
```

## Perform the following preprocess steps on the above  df_survived dataframe
#### 1. Add 2 new columns first_name and second_name by splitting the Name column on comma(,)
#### 2. Remove the prefixes from the second name( ex Mr. , Dr. etc)
#### 3. Trim/Strip all the white spaces in above 2 columns( no space before and after the string in each cell)
#### Your output should look like the below value in last 2 columns
|Name|first_name|second_name|
|---|---|---|
|Daly, Mr. Peter Denis|Daly|Peter Denis|

```python
split_data1=df_survived["Name"].str.split(",")
data1=split_data1.to_list()
df_survived["first_name"]=data1[0][0]
df_survived["second_name"]=data1[0][1]
df_survived["second_name"]=df_survived["second_name"].replace(['Mr. ','Dr. '],'', regex=True)
df_survived
```

## Perform the following additional preprocess steps on the above df_survived dataframe
#### 1. Add a new column economy_group which contain values "Rich" or "Not Rich" depend on whether the Pclass is 1 or not respectively
#### 2. Add a new column age_group which contain values  "old" if age is more than 50,  "child" if age is less than 15 otherwise  "young" 
#### 3. Add a new column name_length which contain the string length for Name column


```python
def new_column(row):
    if row['Age']>50:
        return "old"
    elif row['Age']<15:
        return "child"
    else:
        return "young"
df_survived['Economy_Group']=df_survived['Pclass'].apply(lambda x: 'Rich' if x==1 else 'Not Rich')
df_survived['Age_Group']=df_survived.apply(new_column,axis=1)
df_survived['Name_length']=df_survived['Name'].apply(lambda x: len(x))

df_survived
```

## Create a new dataframe(df_kpi) that calculate the following KPIs by grouping (Pclass, age_group and economy_group) from the above df_survived dataframe
#### 1. total records count
#### 2. total survivors count
#### 3. avg fare
#### 4. maximum fare
#### 5. minimum fare
#### 6. total fare

```python
#df_survived['Survived']=
df_kpi=df_survived.groupby(['Pclass','Age_Group','Economy_Group']).agg({'Survived':lambda x:x.eq(1).sum(),
                                                                   'Fare':['max','min','mean','sum']})

df_kpi.reset_index()
  

```

## Merge df_survived and df_kpi on Pclass , age_group and economy_group columns

```python
mergedata=pd.merge(df_survived,df_kpi,on=['Pclass','Age_Group','Economy_Group'])
mergedata
```
